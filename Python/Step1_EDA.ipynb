```python
# Import modules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

%matplotlib inline
```

```python
# Load CSV file

df = pd.read_csv(os.path.join('..', 'Resources', 'Source', 'Source_file.csv'))
df
```

```python
# Check data types

df.dtypes
```

```python
# Gender: Histogram, and unique values

plt.hist(df['Gender'])
plt.title('Gender')
plt.show()
df.Gender.value_counts()
```

```python
# Age: Histogram, and min value.

plt.hist(df['Age'])
plt.title('Age')
plt.show()
print(f'Minimum Age: {df.Age.min()}')
```

```python
# Driving_License: Histogram, and unique values

plt.hist(df['Driving_License'])
plt.title('Driving_License')
plt.show()
df.Driving_License.value_counts()
```

```python
# Previously_Insured: Histogram, and unique values

plt.hist(df['Previously_Insured'])
plt.title('Previously_Insured')
plt.show()
df.Previously_Insured.value_counts()
```

```python
# Vehicle_Age: Histogram, and unique values

plt.hist(df['Vehicle_Age'])
plt.title('Vehicle_Age')
plt.show()
df.Vehicle_Age.value_counts()
```

```python
# Vehicle_Damage: Histogram, and unique values

plt.hist(df['Vehicle_Damage'])
plt.title('Vehicle_Damage')
plt.show()
df.Vehicle_Damage.value_counts()
```

```python
# Annual_Premium: Histogram, mean, and median

plt.hist(df['Annual_Premium'])
plt.title('Annual_Premium')
plt.show()
print(f'Mean: {df.Annual_Premium.mean()}, Median: {df.Annual_Premium.median()}')
```

```python
# Annual_Premium: Histogram, mean, and median for lower amounts

low_premium_df = df.loc[df['Annual_Premium']<=60000,]
plt.hist(low_premium_df['Annual_Premium'])
plt.title('Low Premium')
plt.show()
print(f'Mean: {low_premium_df.Annual_Premium.mean()}, Median: {low_premium_df.Annual_Premium.median()}')
```

```python
# Annual_Premium: Histogram, mean, and median for medium amounts

medium_premium_df = df.loc[(df['Annual_Premium']>60000)&(df['Annual_Premium']<=120000),]
plt.hist(medium_premium_df['Annual_Premium'])
plt.title('Medium Premium')
plt.show()
print(f'Mean: {medium_premium_df.Annual_Premium.mean()}, Median: {medium_premium_df.Annual_Premium.median()}')
```

```python
# Annual_Premium: Histogram, mean, and median for higher amounts

high_premium_df = df.loc[df['Annual_Premium']>120000,]
plt.hist(high_premium_df['Annual_Premium'])
plt.title('High Premium')
plt.show()
print(f'Mean: {high_premium_df.Annual_Premium.mean()}, Median: {high_premium_df.Annual_Premium.median()}')
```

```python
# Number of data points in each spectrum

print(f'''Data points for lower amounts of premium (<=60000): {len(low_premium_df)}
Data points for lower amounts of premium (60000< <=120000): {len(medium_premium_df)}
Data points for lower amounts of premium (12000<): {len(high_premium_df)}''')
```

```python
# Policy_Sales_Channel: Histogram, and number of unique values

plt.hist(df['Policy_Sales_Channel'])
plt.title('Policy_Sales_Channel')
plt.show()
print(f'Number of unique values: {df.Policy_Sales_Channel.value_counts().count()}')
```

```python
# Vintage: Histogram, and unique values

plt.hist(df['Vintage'])
plt.title('Vintage')
plt.show()
print(f'''Mean: {df.Vintage.mean()}, Median: {df.Vintage.median()}
Max: {df.Vintage.max()}, Min: {df.Vintage.min()}''')
```
