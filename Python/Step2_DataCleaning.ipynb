```python
# Import modules

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import os
```

```python
# Load CSV file

df = pd.read_csv(os.path.join('..', 'Resources', 'Source', 'Source_file.csv'))
df.head(1)
```

```python
# Drop id

df = df.drop(columns=['id'])
df.head(5)
```

```python
# Encode categorical features

encoded_df = df.copy()

le = LabelEncoder()
encoded_df['Gender'] = le.fit_transform(encoded_df['Gender'])
encoded_df['Vehicle_Age'] = le.fit_transform(encoded_df['Vehicle_Age'])
encoded_df['Vehicle_Damage'] = le.fit_transform(encoded_df['Vehicle_Damage'])

encoded_df
```

```python
# Filter out Driving_Licence = 0, and then drop the column

filtered_df = encoded_df[encoded_df['Driving_License'] == 1]
filtered_df = filtered_df.drop(columns=['Driving_License'])
filtered_df = filtered_df.reset_index().drop(columns='index')
filtered_df
```

```python
# Check for nulls

filtered_df.isnull().sum()
```

```python
# Export DataFrame to CSV file

filtered_df.to_csv(os.path.join('..', 'Resources', 'Created_Files', 'Cleaned_df.csv'), index=False)
```
